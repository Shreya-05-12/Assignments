```python
import pandas as pd
df=pd.read_csv(r"C:\Users\shrey\Downloads\Day_9_banking_data.csv")
df
```

```python
import matplotlib.pyplot as plt


account_type_summary = df.groupby("Account_Type")["Transaction_Amount"].sum()
plt.figure(figsize=(8, 6))
account_type_summary.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title(" Total Transaction Amount per Account Type", fontsize=16)
plt.xlabel("Account Type", fontsize=12)
plt.ylabel("Total Transaction Amount", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
