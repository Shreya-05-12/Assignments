```python
sns.barplot(x='Sales',y='Region',data=df,color='red',estimator='mean')
plt.show()
plt.tight_layout()
```

```python
sns.scatter(x='Marketing_spend',y='Sales',data=df,color='red',estimator='mean')
plt.show()
```

```python
sns.boxplot(x=df['Effectiveness'],y=df['Age_Group'],color='blue')
```

```python
sns.lineplot(x=["Trial_Period"], y=["Sales"], hue=["Product"],palette=["green", "orange", "blue"])
plt.show()
```

```python
sns.lineplot(x="Trial_Period", y="Sales", data=df, palette=["green", "orange", "blue"])
plt.title("Sales Trend by Product Over Trial Periods", fontsize=16)
plt.xlabel("Trial Period", fontsize=12)
plt.ylabel("Sales", fontsize=12)
plt.show()
```
